```python
from model import Model

# Number of times to run the simulation
num_simulations = 50
total_captured_percentage = 0

for _ in range(num_simulations):
    # Create a new instance of the model with 1000 initial value
    mod = Model(1000)
    mod.run_simulation()
    total_captured_percentage += (mod.trapped / mod.num_molecules) * 100

average_captured_percentage = total_captured_percentage / num_simulations
print("Average Captured Percentage:", average_captured_percentage)
```
